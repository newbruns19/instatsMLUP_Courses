## Importing modules

```python
# General visualisation and data manipulation modules
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
import pandas as pd

# Classification modules
from sklearn.datasets import make_circles, make_classification, make_moons
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

# Preprocessing modules
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

# Dimentionality reduction
from sklearn.decomposition import PCA

# Clustering
from sklearn.cluster import KMeans

# Regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.inspection import permutation_importance
from sklearn.metrics import r2_score, mean_squared_error

# Pipelines 
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Datasets
from sklearn.datasets import load_iris
```

## Example of the learning algorithms available in sklearn

```python
names = [
    "Nearest Neighbors",
    "Linear SVM",
    "RBF SVM",
    "Gaussian Process",
    "Decision Tree",
    "Random Forest",
    "Neural Net",
    "AdaBoost",
    "Naive Bayes",
    "QDA",
]

classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025, random_state=42),
    SVC(gamma=2, C=1, random_state=42),
    GaussianProcessClassifier(1.0 * RBF(1.0), random_state=42),
    DecisionTreeClassifier(max_depth=5, random_state=42),
    RandomForestClassifier(
        max_depth=5, n_estimators=10, max_features=1, random_state=42
    ),
    MLPClassifier(alpha=1, max_iter=1000, random_state=42),
    AdaBoostClassifier(random_state=42),
    GaussianNB(),
    QuadraticDiscriminantAnalysis(),
]

X, y = make_classification(
    n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1
)
rng = np.random.RandomState(2)
X += 2 * rng.uniform(size=X.shape)
linearly_separable = (X, y)

datasets = [
    make_moons(noise=0.3, random_state=0),
    make_circles(noise=0.2, factor=0.5, random_state=1),
    linearly_separable,
]

figure = plt.figure(figsize=(27, 9))
i = 1
# iterate over datasets
for ds_cnt, ds in enumerate(datasets):
    # preprocess dataset, split into training and test part
    X, y = ds
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.4, random_state=42
    )

    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

    # just plot the dataset first
    cm = plt.cm.RdBu
    cm_bright = ListedColormap(["#FF0000", "#0000FF"])
    ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
    if ds_cnt == 0:
        ax.set_title("Input data")
    # Plot the training points
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors="k")
    # Plot the testing points
    ax.scatter(
        X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6, edgecolors="k"
    )
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xticks(())
    ax.set_yticks(())
    i += 1

    # iterate over classifiers
    for name, clf in zip(names, classifiers):
        ax = plt.subplot(len(datasets), len(classifiers) + 1, i)

        clf = make_pipeline(StandardScaler(), clf)
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)
        DecisionBoundaryDisplay.from_estimator(
            clf, X, cmap=cm, alpha=0.8, ax=ax, eps=0.5
        )

        # Plot the training points
        ax.scatter(
            X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors="k"
        )
        # Plot the testing points
        ax.scatter(
            X_test[:, 0],
            X_test[:, 1],
            c=y_test,
            cmap=cm_bright,
            edgecolors="k",
            alpha=0.6,
        )

        ax.set_xlim(x_min, x_max)
        ax.set_ylim(y_min, y_max)
        ax.set_xticks(())
        ax.set_yticks(())
        if ds_cnt == 0:
            ax.set_title(name)
        ax.text(
            x_max - 0.3,
            y_min + 0.3,
            ("%.2f" % score).lstrip("0"),
            size=15,
            horizontalalignment="right",
        )
        i += 1

plt.tight_layout()
plt.show()
```

## Dimensionality reduction

```python
# Environmental data: [temp_mean, temp_range, precip_annual, precip_seasonality, 
#                     elevation, soil_ph, solar_radiation]
environmental_data = np.array([
    [15.2, 12.4, 1200, 0.3, 450, 6.2, 18.5],   # Site 1
    [18.7, 10.1, 800, 0.5, 1200, 5.8, 16.2],   # Site 2
    [12.1, 15.2, 1500, 0.2, 200, 7.1, 19.1],   # Site 3
    [20.3, 8.7, 600, 0.7, 800, 6.0, 15.8],     # Site 4
    [14.8, 13.1, 1100, 0.4, 600, 6.5, 17.9],   # Site 5
    [22.1, 7.2, 450, 0.8, 1500, 5.5, 14.2]     # Site 6
])

# Apply PCA to identify main environmental gradients
pca = PCA(n_components=3)
env_pca = pca.fit_transform(environmental_data)

print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Cumulative variance explained:", pca.explained_variance_ratio_.cumsum())

# Examine component loadings to interpret ecological meaning
feature_names = ['temp_mean', 'temp_range', 'precip_annual', 'precip_seasonality', 
                'elevation', 'soil_ph', 'solar_radiation']
for i, component in enumerate(pca.components_[:2]):  # First 2 components
    print(f"\nPC{i+1} loadings:")
    for feature, loading in zip(feature_names, component):
        print(f"  {feature}: {loading:.3f}")
```

## Clustering

```python
# Environmental data: [temperature, precipitation, elevation, soil_pH, canopy_cover]
environmental_data = np.array([
    [15.2, 1200, 450, 6.2, 85],    # Temperate forest
    [18.7, 800, 1200, 5.8, 45],    # Mountain grassland  
    [12.1, 1500, 200, 7.1, 90],    # Wetland forest
    [20.3, 600, 800, 6.0, 30],     # Dry grassland
    [16.1, 1100, 400, 6.4, 80],    # Mixed forest
    [19.2, 650, 850, 5.9, 25],     # Scrubland
    [13.8, 1400, 250, 6.8, 88],    # Riparian forest
    [21.1, 550, 900, 5.7, 20],     # Desert edge
    [14.9, 1250, 350, 6.6, 82],    # Deciduous forest
    [17.8, 750, 950, 6.1, 35]      # Alpine meadow
])

# Standardize environmental variables
scaler = StandardScaler()
env_scaled = scaler.fit_transform(environmental_data)

# Apply K-means clustering to identify habitat types
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
habitat_clusters = kmeans.fit_predict(env_scaled)

print("Site cluster assignments:", habitat_clusters)
print("Cluster centers (standardized):")
for i, center in enumerate(kmeans.cluster_centers_):
    print(f"Habitat type {i+1}: {center}")

# Calculate within-cluster sum of squares (inertia)
print(f"Within-cluster sum of squares: {kmeans.inertia_:.2f}")

# Interpret clusters by examining original data
for cluster_id in range(3):
    sites_in_cluster = environmental_data[habitat_clusters == cluster_id]
    print(f"\nHabitat type {cluster_id + 1} characteristics:")
    print(f"  Mean temperature: {sites_in_cluster[:, 0].mean():.1f}°C")
    print(f"  Mean precipitation: {sites_in_cluster[:, 1].mean():.0f}mm")
    print(f"  Mean elevation: {sites_in_cluster[:, 2].mean():.0f}m")
```

## Regression

```python
# Complex ecological dataset: predicting species richness from multiple factors
np.random.seed(42)
n_sites = 200

# Generate correlated environmental variables
temperature = np.random.normal(15, 5, n_sites)
precipitation = np.random.normal(1000, 300, n_sites)
elevation = np.random.uniform(0, 2000, n_sites)
habitat_diversity = np.random.uniform(0.1, 1.0, n_sites)
human_disturbance = np.random.uniform(0, 1, n_sites)
fragment_size = np.random.uniform(1, 1000, n_sites)

# Complex non-linear relationship for species richness
# Include interactions and thresholds common in ecology
species_richness = (
    2 * temperature + 
    0.01 * precipitation + 
    -0.002 * elevation + 
    30 * habitat_diversity + 
    -25 * human_disturbance + 
    0.05 * fragment_size +
    # Interaction terms
    0.1 * temperature * habitat_diversity +
    -0.02 * human_disturbance * fragment_size +
    # Threshold effect for fragment size
    10 * (fragment_size > 100).astype(int) +
    np.random.normal(0, 5, n_sites)
)

# Ensure positive species richness
species_richness = np.maximum(species_richness, 0)

# Create DataFrame
eco_data = pd.DataFrame({
    'temperature': temperature,
    'precipitation': precipitation,
    'elevation': elevation,
    'habitat_diversity': habitat_diversity,
    'human_disturbance': human_disturbance,
    'fragment_size': fragment_size,
    'species_richness': species_richness
})

# Prepare features and target
X = eco_data.drop('species_richness', axis=1)
y = eco_data['species_richness']

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Hyperparameter tuning with GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 15, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(
    rf, param_grid, cv=5, scoring='r2', n_jobs=-1, verbose=1
)

print("Performing hyperparameter tuning...")
grid_search.fit(X_train, y_train)

# Best model
best_rf = grid_search.best_estimator_
print(f"Best parameters: {grid_search.best_params_}")

# Make predictions
y_pred = best_rf.predict(X_test)

# Evaluate performance
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"\nModel Performance:")
print(f"R² Score: {r2:.3f}")
print(f"RMSE: {rmse:.2f} species")

# Feature importance analysis
feature_importance = best_rf.feature_importances_
feature_names = X.columns

# Permutation importance (more reliable)
perm_importance = permutation_importance(
    best_rf, X_test, y_test, n_repeats=10, random_state=42
)

# Create importance DataFrame
importance_df = pd.DataFrame({
    'feature': feature_names,
    'importance_builtin': feature_importance,
    'importance_permutation': perm_importance.importances_mean,
    'importance_std': perm_importance.importances_std
})

importance_df = importance_df.sort_values('importance_permutation', ascending=False)
print(f"\nFeature Importance (Permutation-based):")
for _, row in importance_df.iterrows():
    print(f"{row['feature']}: {row['importance_permutation']:.3f} ± {row['importance_std']:.3f}")

# Visualization
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Predicted vs Actual
axes[0].scatter(y_test, y_pred, alpha=0.6)
axes[0].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
axes[0].set_xlabel('Actual Species Richness')
axes[0].set_ylabel('Predicted Species Richness')
axes[0].set_title(f'Predicted vs Actual (R² = {r2:.3f})')

# Feature importance
axes[1].barh(importance_df['feature'], importance_df['importance_permutation'])
axes[1].set_xlabel('Permutation Importance')
axes[1].set_title('Feature Importance')

# Residuals
residuals = y_test - y_pred
axes[2].scatter(y_pred, residuals, alpha=0.6)
axes[2].axhline(y=0, color='r', linestyle='--')
axes[2].set_xlabel('Predicted Species Richness')
axes[2].set_ylabel('Residuals')
axes[2].set_title('Residual Plot')

plt.tight_layout()
plt.show()
```

## Pipelines

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Convert target numbers to species names for better interpretability
species_names = iris.target_names
y_named = [species_names[target] for target in y]

print("Iris Dataset Information:")
print("=" * 30)
print(f"Number of samples: {X.shape[0]}")
print(f"Number of features: {X.shape[1]}")
print(f"Feature names: {iris.feature_names}")
print(f"Species: {list(species_names)}")
print(f"Samples per species: {np.bincount(y)}")

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y_named, test_size=0.3, random_state=42, stratify=y_named
)
```

```python
# Define pipelines for each algorithm
pipelines = {
    'Decision Tree': Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', DecisionTreeClassifier(random_state=42))
    ]),
    'SVM': Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', SVC(random_state=42))
    ]),
    'KNN': Pipeline([
        ('scaler', StandardScaler()),
        ('classifier', KNeighborsClassifier())
    ])
}
```

```python
# Evaluate each pipeline
results = {}
feature_names = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

print("Species Classification Results:")
print("=" * 50)

for name, pipeline in pipelines.items():
    # Cross-validation scores
    cv_scores = cross_val_score(pipeline, X_train, y_train, cv=5, scoring='accuracy')
    
    # Fit and predict
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_pred)
    
    results[name] = {
        'cv_mean': cv_scores.mean(),
        'cv_std': cv_scores.std(),
        'test_accuracy': test_accuracy,
        'pipeline': pipeline
    }
    
    print(f"\n{name}:")
    print(f"  Cross-validation accuracy: {cv_scores.mean():.3f} ± {cv_scores.std():.3f}")
    print(f"  Test accuracy: {test_accuracy:.3f}")
```

```python
# Select best model
best_model_name = max(results.keys(), key = lambda k: results[k]['test_accuracy'])
best_pipeline = results[best_model_name]['pipeline']

print(f"\nBest Model: {best_model_name}")
print(f"Best Test Accuracy: {results[best_model_name]['test_accuracy']:.3f}")

# Detailed classification report for best model
y_pred_best = best_pipeline.predict(X_test)
print(f"\nClassification Report for {best_model_name}:")
print(classification_report(y_test, y_pred_best))
```
