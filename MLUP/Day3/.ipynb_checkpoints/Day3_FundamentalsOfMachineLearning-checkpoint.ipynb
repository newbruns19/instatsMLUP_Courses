## Importing main modules

```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, roc_curve, auc, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
```

## Functions used

```python
def euclidean_distance(x1, x2):
    """Calculate Euclidean distance between two points"""
    return np.sqrt(np.sum((x1 - x2) ** 2))

def knn_classify_numpy(X_train, y_train, X_test, k=3):
    """
    K-Nearest Neighbors classification using pure NumPy
    Based on the implementation pattern from search results[2]
    """
    predictions = []
    
    for test_point in X_test:
        # Calculate distances to all training points
        distances = []
        for i, train_point in enumerate(X_train):
            dist = euclidean_distance(test_point, train_point)
            distances.append((dist, y_train[i]))
        
        # Sort by distance and get k nearest neighbors
        distances.sort(key = lambda x: x[0])
        k_nearest = distances[:k]
        
        # Get the most common class among k neighbors
        neighbor_classes = [neighbor[1] for neighbor in k_nearest]
        prediction = max(set(neighbor_classes), key=neighbor_classes.count)
        predictions.append(prediction)
    
    return np.array(predictions)
```

## Same Exercises

```python
# Create synthetic bird measurement data
np.random.seed(42)
n_samples = 300

# Three bird species: Robin (0), Sparrow (1), Cardinal (2)
species_labels = []
wing_length = []
body_mass = []
beak_length = []
tail_length = []

# Robin characteristics (smaller, lighter)
for i in range(100):
    species_labels.append(0)
    wing_length.append(np.random.normal(12.5, 1.2))
    body_mass.append(np.random.normal(16.5, 2.1))
    beak_length.append(np.random.normal(1.8, 0.3))
    tail_length.append(np.random.normal(5.2, 0.8))

# Sparrow characteristics (medium)
for i in range(100):
    species_labels.append(1)
    wing_length.append(np.random.normal(9.8, 0.9))
    body_mass.append(np.random.normal(24.2, 3.2))
    beak_length.append(np.random.normal(1.1, 0.2))
    tail_length.append(np.random.normal(4.5, 0.6))

# Cardinal characteristics (larger, heavier)
for i in range(100):
    species_labels.append(2)
    wing_length.append(np.random.normal(15.2, 1.5))
    body_mass.append(np.random.normal(45.8, 4.8))
    beak_length.append(np.random.normal(2.3, 0.4))
    tail_length.append(np.random.normal(7.1, 1.1))

# Create DataFrame
bird_data = pd.DataFrame({
    'wing_length_cm': wing_length,
    'body_mass_g': body_mass,
    'beak_length_cm': beak_length,
    'tail_length_cm': tail_length,
    'species': species_labels
})

print("Bird Dataset Preview:")
print(bird_data.head())
print(f"\nDataset shape: {bird_data.shape}")
```

```python
# Prepare the data
X = bird_data[['wing_length_cm', 'body_mass_g', 'beak_length_cm', 'tail_length_cm']].values
y = bird_data['species'].values

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Normalize features (important for KNN as mentioned in search results[5])
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Apply our NumPy KNN
y_pred_numpy = knn_classify_numpy(X_train_scaled, y_train, X_test_scaled, k=5)
accuracy_numpy = accuracy_score(y_test, y_pred_numpy)

print(f"NumPy KNN Accuracy: {accuracy_numpy:.3f}")
```

```python
# Using scikit-learn KNN
knn_sklearn = KNeighborsClassifier(n_neighbors=5)
knn_sklearn.fit(X_train_scaled, y_train)
y_pred_sklearn = knn_sklearn.predict(X_test_scaled)
accuracy_sklearn = accuracy_score(y_test, y_pred_sklearn)

print(f"Scikit-learn KNN Accuracy: {accuracy_sklearn:.3f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_sklearn, target_names=['Robin', 'Sparrow', 'Cardinal']))
```

```python
# Create a second ecological dataset - plant habitat preferences
np.random.seed(123)
n_plants = 250

# Features: soil_ph, moisture_level, sunlight_hours, temperature_avg
# Classes: Forest (0), Grassland (1), Wetland (2)

plant_data = []
plant_labels = []

# Forest plants (acidic soil, moderate moisture, low sunlight)
for i in range(80):
    soil_ph = np.random.normal(5.2, 0.8)
    moisture = np.random.normal(65, 15)
    sunlight = np.random.normal(4.5, 1.2)
    temperature = np.random.normal(18.5, 3.5)
    plant_data.append([soil_ph, moisture, sunlight, temperature])
    plant_labels.append(0)

# Grassland plants (neutral soil, low moisture, high sunlight)
for i in range(85):
    soil_ph = np.random.normal(6.8, 0.6)
    moisture = np.random.normal(35, 12)
    sunlight = np.random.normal(8.2, 1.8)
    temperature = np.random.normal(22.1, 4.2)
    plant_data.append([soil_ph, moisture, sunlight, temperature])
    plant_labels.append(1)

# Wetland plants (neutral to basic soil, high moisture, moderate sunlight)
for i in range(85):
    soil_ph = np.random.normal(7.5, 0.7)
    moisture = np.random.normal(85, 10)
    sunlight = np.random.normal(6.1, 1.5)
    temperature = np.random.normal(20.3, 3.8)
    plant_data.append([soil_ph, moisture, sunlight, temperature])
    plant_labels.append(2)

X_plants = np.array(plant_data)
y_plants = np.array(plant_labels)
```

```python
# Train KNN on plant data
X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(X_plants, y_plants, test_size=0.25, random_state=42)

scaler_plants = StandardScaler()
X_train_p_scaled = scaler_plants.fit_transform(X_train_p)
X_test_p_scaled = scaler_plants.transform(X_test_p)

knn_plants = KNeighborsClassifier(n_neighbors=7)
knn_plants.fit(X_train_p_scaled, y_train_p)
y_pred_plants = knn_plants.predict(X_test_p_scaled)

print(f"\nPlant Habitat KNN Accuracy: {accuracy_score(y_test_p, y_pred_plants):.3f}")
```

```python
# Dataset 1: Pollinator Activity Classification
np.random.seed(456)
n_observations = 400

# Features: temperature, humidity, wind_speed, flower_density
# Binary classification: Active (1) or Inactive (0) pollinator behavior

pollinator_data = []
pollinator_labels = []

# Active conditions (warm, moderate humidity, low wind, high flower density)
for i in range(200):
    temperature = np.random.normal(24.5, 4.2)
    humidity = np.random.normal(55, 12)
    wind_speed = np.random.normal(2.1, 1.8)
    flower_density = np.random.normal(85, 15)
    
    # Add some noise to make classification challenging
    noise_factor = np.random.normal(0, 0.1)
    activity_score = (temperature/30) + (1-humidity/100) + (1-wind_speed/10) + (flower_density/100) + noise_factor
    
    pollinator_data.append([temperature, humidity, wind_speed, flower_density])
    pollinator_labels.append(1 if activity_score > 1.8 else 0)

# Inactive conditions (cool, high humidity, high wind, low flower density)
for i in range(200):
    temperature = np.random.normal(15.2, 3.8)
    humidity = np.random.normal(78, 18)
    wind_speed = np.random.normal(6.5, 2.2)
    flower_density = np.random.normal(35, 20)
    
    noise_factor = np.random.normal(0, 0.1)
    activity_score = (temperature/30) + (1-humidity/100) + (1-wind_speed/10) + (flower_density/100) + noise_factor
    
    pollinator_data.append([temperature, humidity, wind_speed, flower_density])
    pollinator_labels.append(1 if activity_score > 1.8 else 0)

X_poll = np.array(pollinator_data)
y_poll = np.array(pollinator_labels)

print(f"Pollinator dataset - Active: {sum(y_poll)}, Inactive: {len(y_poll) - sum(y_poll)}")
```

```python
# Split data
X_train_dt, X_test_dt, y_train_dt, y_test_dt = train_test_split(X_poll, y_poll, test_size=0.3, random_state=42)

# Train Decision Tree
dt_classifier = DecisionTreeClassifier(random_state=42, max_depth=5, min_samples_split=10)
dt_classifier.fit(X_train_dt, y_train_dt)

# Make predictions
y_pred_dt = dt_classifier.predict(X_test_dt)
y_pred_proba_dt = dt_classifier.predict_proba(X_test_dt)[:, 1]  # Probabilities for positive class

# Compute Confusion Matrix
cm = confusion_matrix(y_test_dt, y_pred_dt)
print("Confusion Matrix:")
print(cm)

# Calculate performance metrics as described in search results[3]
tn, fp, fn, tp = cm.ravel()

# True Positive Rate (Sensitivity/Recall)
tpr = tp / (tp + fn)
print(f"True Positive Rate (TPR): {tpr:.3f}")

# False Positive Rate 
fpr = fp / (fp + tn)
print(f"False Positive Rate (FPR): {fpr:.3f}")

# Accuracy
accuracy = (tp + tn) / (tp + tn + fp + fn)
print(f"Accuracy: {accuracy:.3f}")

# Alternative accuracy calculation
accuracy_sklearn = accuracy_score(y_test_dt, y_pred_dt)
print(f"Accuracy (sklearn): {accuracy_sklearn:.3f}")
```

```python
# Calculate ROC curve
fpr_roc, tpr_roc, thresholds = roc_curve(y_test_dt, y_pred_proba_dt)

# Calculate AUC
auc_score = auc(fpr_roc, tpr_roc)
# Alternative AUC calculation
auc_sklearn = roc_auc_score(y_test_dt, y_pred_proba_dt)

print(f"Area Under ROC Curve (AUC): {auc_score:.3f}")
print(f"AUC (sklearn): {auc_sklearn:.3f}")

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_roc, tpr_roc, color='darkorange', lw=2, label=f'ROC curve (AUC = {auc_score:.3f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random classifier')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Pollinator Activity Classification')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

# Display detailed confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Inactive', 'Active'], 
            yticklabels=['Inactive', 'Active'])
plt.title('Confusion Matrix for Pollinator Activity')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

```python
# Dataset 2: Soil Quality Assessment (Multi-class)
np.random.seed(789)
n_soil_samples = 350

# Features: organic_matter, nitrogen, phosphorus, ph_level
# Classes: Poor (0), Good (1), Excellent (2)

soil_data = []
soil_labels = []

# Poor soil quality
for i in range(120):
    organic_matter = np.random.normal(1.8, 0.6)
    nitrogen = np.random.normal(0.08, 0.03)
    phosphorus = np.random.normal(12, 8)
    ph_level = np.random.normal(5.2, 1.2)
    soil_data.append([organic_matter, nitrogen, phosphorus, ph_level])
    soil_labels.append(0)

# Good soil quality
for i in range(115):
    organic_matter = np.random.normal(3.5, 0.8)
    nitrogen = np.random.normal(0.15, 0.04)
    phosphorus = np.random.normal(25, 10)
    ph_level = np.random.normal(6.5, 0.8)
    soil_data.append([organic_matter, nitrogen, phosphorus, ph_level])
    soil_labels.append(1)

# Excellent soil quality
for i in range(115):
    organic_matter = np.random.normal(5.2, 1.1)
    nitrogen = np.random.normal(0.25, 0.05)
    phosphorus = np.random.normal(45, 12)
    ph_level = np.random.normal(7.1, 0.6)
    soil_data.append([organic_matter, nitrogen, phosphorus, ph_level])
    soil_labels.append(2)

X_soil = np.array(soil_data)
y_soil = np.array(soil_labels)

# Train multi-class decision tree
X_train_soil, X_test_soil, y_train_soil, y_test_soil = train_test_split(X_soil, y_soil, test_size=0.3, random_state=42)

dt_soil = DecisionTreeClassifier(random_state=42, max_depth=6)
dt_soil.fit(X_train_soil, y_train_soil)
y_pred_soil = dt_soil.predict(X_test_soil)

# Multi-class confusion matrix
cm_soil = confusion_matrix(y_test_soil, y_pred_soil)
print("\nSoil Quality Confusion Matrix:")
print(cm_soil)

# Calculate per-class metrics
from sklearn.metrics import classification_report
print("\nDetailed Classification Report:")
print(classification_report(y_test_soil, y_pred_soil, target_names=['Poor', 'Good', 'Excellent']))

# Visualize multi-class confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_soil, annot=True, fmt='d', cmap='Greens',
            xticklabels=['Poor', 'Good', 'Excellent'],
            yticklabels=['Poor', 'Good', 'Excellent'])
plt.title('Confusion Matrix for Soil Quality Classification')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```
